```python
import tweepy
import json
import datetime
import time
import pandas as pd
import pathlib
import glob
import os

# API access details - get yours from Twitter
twitter_auth={'API key':'******',
             'API secret key':'******',
             'Access token':'******',
             'Access token secret':'******'}
#Authenticate to Twitter api using our account keys
auth_details = tweepy.OAuthHandler(twitter_auth['API key'],twitter_auth['API secret key'] )
auth_details.set_access_token(twitter_auth['Access token'], twitter_auth['Access token secret'])
 
#Create connection to the Twitter api
api = tweepy.API(auth_details)

#Loop and sleep to download tweets
for x in range(0,48):
    tweet_array = []
    tweets = api.home_timeline(count=100)
    for tweet in tweets:
        t = {"text" : tweet.text, "likes": tweet.favorite_count, "retweets": tweet.retweet_count }
        tweet_array.append(t)
        d = datetime.datetime.today().strftime("%Y-%m-%d-%H-%M-%S")
        f = open('tweets-' + d + ".json", 'w')
        f.write(json.dumps(tweet_array))
        f.close()
    print("Sleeping")
    time.sleep(60 * 15)
#exporting csv    
tweets = []

for file in glob.glob('*.json'):
    with open(file) as json_data:
        data = json.load(json_data)
        for tweet in data:
            tweets.append(tweet)

tweet_df=pd.DataFrame(tweets)
tweet_df.to_csv('tweet_1_df.csv',index=False)

#removing json files
for f in glob.glob("*.json"):
    os.remove(f)
```
