```python
import pandas as pd
import datetime as dt
from twitterscraper.query import query_tweets_once

#query function to output query format for a user for a given time duration
def query_format_output(user,days=3):
    
    #today-date
    today=dt.date.today()
    
    #the days tweets we want
    since=today-dt.timedelta(days=days)
    until=since+dt.timedelta(days=1)
    
    #query format
    query='(from:{}) until:{} since:{}'.format(user,until,since)
    
    return(query)

#function to get tweets of a pirticular user few days back

def get_user_tweets(user,days=3):
    
    #query for a user,few days back
    query=query_format_output(user,days)
    
    #tweets of the user
    tweets=query_tweets_once(query,lang='en')
    
    #data frame of the user
    df=pd.DataFrame(t.__dict__ for t in tweets)
    return(df)

#making data frame for list of users

def get_tweets_of_all_users(users,data_path='E:/twitter_data/twitter_users_data/',days=3):
    
    today=dt.date.today()
    
    #empty dataframe
    df=pd.DataFrame()
    
    for user in users:
        
        #dataframe of user
        df_user=get_user_tweets(user,days)
        
        #concating to main dataframe
        df=pd.concat([df,df_user])
       
    #Exporting csvfile
    df.to_csv(data_path+'tweet_{}.csv'.format(str(today-dt.timedelta(days=0))),index=False)


#location of the file
data_path='E:/twitter_data/twitter_users_data/'

#list of users
with open(data_path+'users.txt','r') as f:
        users=f.read()
users=users.split(',')[:-1]

#no of days
days=3

get_tweets_of_all_users(users,data_path,days)
```
