```python
#importing helper functions
from ipynb.fs.full.data_cleaning import get_tweet_csv,get_bert_emb_csv
from ipynb.fs.full.feature_eng_funcs import engage_score,get_followers_dict
```

```python
#retriveing the daily twitter data collected
df_tweets=get_tweet_csv()
df_pools=get_bert_emb_csv()
df_pools.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
#getting the number of followers for each user
user_folls=get_followers_dict(df_tweets)
```

```python
#making followers column using the user_folls dict
df_tweets['followers']=df_tweets['screen_name'].apply(lambda name:user_folls[name])
```

```python
#calculating the score
df_tweets['score']=df_tweets.apply(lambda tweet:engage_score(tweet.likes,tweet.retweets,tweet.replies,tweet.followers),axis=1)
```

```python
df_pools
```

```python
df_tweets.columns
```

```python
#joined both data frames into one
df=df_pools.join(df_tweets)
```

```python
#just did this step to save the dataframe so that i dont lose it is 200MB in size
import pandas as pd
data_path='E:/twitter_data/twitter_users_data/'
df.to_csv(data_path+'df.csv',index=False)
df=pd.read_csv(data_path+'df.csv')
```

```python
#importing xgboost and randomforrest
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb

from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
#droping all rows except bert embeddings
X_train=df.drop(['screen_name', 'username', 'user_id', 'tweet_id', 'tweet_url',
       'timestamp', 'timestamp_epochs', 'text', 'text_html', 'links',
       'hashtags', 'has_media', 'img_urls', 'video_url', 'likes', 'retweets',
       'replies', 'is_replied', 'is_reply_to', 'parent_tweet_id',
       'reply_to_users', 'followers', 'score'],axis=1).iloc[:-200]
y_train=df['score'].iloc[:-200]
X_test=df.drop(['screen_name', 'username', 'user_id', 'tweet_id', 'tweet_url',
       'timestamp', 'timestamp_epochs', 'text', 'text_html', 'links',
       'hashtags', 'has_media', 'img_urls', 'video_url', 'likes', 'retweets',
       'replies', 'is_replied', 'is_reply_to', 'parent_tweet_id',
       'reply_to_users', 'followers', 'score'],axis=1).iloc[-200:]
y_test=df['score'].iloc[-200:]
```

```python
rfr=RandomForestRegressor()
```

```python
#this is the direct train with default parameters
rfr.fit(X_train,y_train)#took 10 min
```

```python
pred_rfr=rfr.predict(X_test)
```

```python
import numpy as np
```

```python
print(mean_absolute_error(y_test,pred_rfr))
print(mean_squared_error(y_test,pred_rfr))
print(np.sqrt(mean_squared_error(y_test,pred_rfr)))
#it is fine with this error for score, well error under 0.5 almost explains the quality of the tweet. its pretty darn good 😉
```

```python
xbr=xgb.XGBRegressor()
```

```python
xbr.fit(X_train,y_train)
```

```python
pred_xbr=xbr.predict(X_test)
```

```python
print(mean_absolute_error(y_test,pred_xbr))
print(mean_squared_error(y_test,pred_xbr))
print(np.sqrt(mean_squared_error(y_test,pred_xbr)))
#yessssss rfr was gooooooood 🤩
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#looking at how well it worked
plt.figure(figsize=(16,9))
sns.scatterplot(x=y_test,y=pred_rfr)
sns.scatterplot(x=y_test,y=pred_xbr,alpha=0.5)
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(x=y_test,y=pred_rfr)
sns.lineplot(x=y_test,y=pred_xbr,alpha=0.5)
```
